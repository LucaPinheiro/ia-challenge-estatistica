```python
import pandas as pd
import matplotlib.pyplot as plt

# Dicionário de faturamento
dict_faturamento = {
    'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'valor': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}
```

```python
df = pd.DataFrame.from_dict(dict_faturamento)
df
```

```python
media_de_vendas = df['valor'].mean()
print(f'A média de vendas é de R$ {media_de_vendas:.2f}')
```

```python
plt.bar(df['data_ref'], df['valor'])
plt.xticks(rotation=45)
plt.figure(figsize=(10, 6))
```

```python
#grafico de linhas
plt.plot(df['data_ref'], df['valor'], marker='o')
plt.xticks(rotation=45)
```
